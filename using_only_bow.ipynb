```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/train.csv', on_bad_lines='skip')

```

```python
df.head()
```

```python
df.shape
```

```python
new_df=df.sample(25000)
```

```python
new_df.shape
```

```python
new_df.isnull().sum()
```

```python
new_df.duplicated().sum()
```

```python
ques_df=new_df[['question1','question2']]
```

```python
ques_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=3000)
```

```python
questions=list(ques_df['question1'])+list(ques_df['question2'])
```

```python
q1_arr,q2_arr=np.split(cv.fit_transform(questions).toarray(),2)
```

```python
temp_df1=pd.DataFrame(q1_arr,index=ques_df.index)
temp_df2=pd.DataFrame(q2_arr,index=ques_df.index)
```

```python
temp_df=pd.concat([temp_df1,temp_df2],axis=1)
```

```python
temp_df.shape
```

```python
temp_df
```

```python
temp_df['is_duplicate']=new_df['is_duplicate']
```

```python
temp_df
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(temp_df.iloc[:,0:-1].values,temp_df.iloc[:,-1].values,test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier()
```

```python
RF.fit(X_train,y_train)
```

```python
y_pred=RF.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from xgboost import XGBClassifier
xgb=XGBClassifier()
```

```python
xgb.fit(X_train,y_train)
```

```python
y_pred1=xgb.predict(X_test)
```

```python
accuracy_score(y_test,y_pred1)
```
