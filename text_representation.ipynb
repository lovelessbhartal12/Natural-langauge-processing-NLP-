```python
import pandas as pd
import numpy as npmb
```

```python
df=pd.DataFrame({'text':['people watch campusx','campusx watch campusx','people watch comment','campusx write comment'],'output':[1,1,0,0]})
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(ngram_range=(2,2))
```

```python
bow=cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
```

```python
cv.transform(['campusx watch and write comment of campusx']).toarray()
```
