```python
import spacy
```

```python
nlp=spacy.load("en_core_web_sm")
```

```python
doc=nlp('i am an enignerr working the google')
```

```python
doc.text
```

```python
doc[0].pos_
```

```python
doc[2].pos_
```

```python
doc[1].tag_
```

```python
spacy.explain('VBP')
```

```python
doc1=nlp(' i read the books about the NLP at the college')
```

```python
for word in doc1:
  print(word.text,"---------->", word.pos_, word.tag_, spacy.explain(word.tag_))
```
